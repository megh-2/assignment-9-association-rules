```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies = pd.read_csv(r"C:\Users\91863\Downloads\my_movies.csv")
movies
```

```python
movies.drop(columns=["V1","V2","V3","V4","V5"],axis=1,inplace=True)
```

```python
movies
```

```python
movies.shape
```

```python
movies.isnull().sum()
```

# support =10% and confidence = 70%

```python
# apriori algo

frequent_itemsets = apriori(movies,min_support=0.1,use_colnames=True)
frequent_itemsets
```

```python
rule=association_rules(frequent_itemsets,metric = "lift",min_threshold=0.7)
rule
```

```python
rule.sort_values("lift",ascending=False)[0:20]
```

```python
rule[rule.lift>1] # good influential value in selecting transacctions
```

```python

# visualization of obtained rule
plt.scatter(rule['support'],rule['confidence'])
plt.xlabel('support')
plt.ylabel('confidence') 
plt.show()
```
